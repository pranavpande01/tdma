```python
import yfinance as yf
import matplotlib.pyplot as plt

def tdma(ticker,duration='1Y',field='Close',period=20):
  print("Details:",ticker,duration,field,period)
  period=period
  field=field
  duration=duration
  ticker=ticker

  msft=yf.Ticker(ticker)
  hist=msft.history(period=duration)

  dates=hist.index.tolist()
  dates=(dates[0:len(dates)-period])
  _dates=[]
  for i in dates:
    _dates.append(i.to_pydatetime().date())
  
  close=[]
  for i in hist[field]:
    close.append(i)
  close.reverse()
  tdma=[]
  x=0

  for i in range(0,len(close)-period):
    tdma.append(sum(close[x:x+period])/period)
    x=x+1
  tdma.reverse()
  close.reverse()
  plt.plot(_dates,tdma)
  plt.plot(_dates,close[period:])
  
```

```python
tdma('msft','1y','Close',5)
```
